```python
#slicing loc vs iloc
import pandas as pd
data = pd.read_csv("../files/cars.csv", sep = ';')
data.head()
```

```python
#25 th row and column name is MPG
data.loc[25,'MPG']
```

```python
data.loc[data.Origin=='US']
```

```python
data.iloc[:]#get all
```

```python
data.iloc[4:6]
```

```python
data.iloc[0,4] #iloc will not select 0 and 4, but 0th row 4th column
```

```python
data.iloc[[0,4]]# now adding one more [] will project 0 and 4 row
```

```python
data.iloc[0:4, 0]  # first 4 rows and 0th column
```

```python
data.iloc[0:4, 0:2]   #first 4 rows and first 2 columns
```
